##### Me resulta interesante en el caso conocer la 'moda' en cada una de las variables, para lo cual importare la herramienta necesaria de python. Esto permitira ver los valores más 'repetidos'. Iteraré el proceso con la mediana que facilitara un entendimiento de la posicion central en la dispersión de valores.

```python
import statistics as stat
from numpy import nanmedian, NaN

# moda y mediana de ambientes
print('La moda de ambientes es de:', stat.mode(new_data.rooms))
print('La mediana de ambientes es de:', np.median(new_data.rooms))

# moda y mediana de dormitorios
print('La moda de dormitorios es de:', stat.mode(new_data.bedrooms))
print('La mediana de dormitorios es de:', np.median(new_data.bedrooms))

# moda y mediana de baños
print('La moda de baños es de:', stat.mode(new_data.bathrooms))
print('La mediana de baños es de:', np.median(new_data.bathrooms))

# moda y mediana de superficie total
print('La moda de superficie total es de:', stat.mode(new_data.surface_total))
print('La mediana de superficie total es de:', np.median(new_data.surface_total))

# moda y mediana de superficie cubierta
print('La moda de superficie cubierta es de:', stat.mode(new_data.surface_covered))
print('La mediana de superficie cubierta es de:', np.median(new_data.surface_covered))

# moda y mediana de precio
print('La moda de precio es de:', stat.mode(new_data.price))
print('La mediana de precio es de:', np.median(new_data.price))
```

##### **MÉTRICAS DE EVALUACIÓN**: 
##### Con el propósito de evaluar correctamente nuestros modelos de machine learning repasaremos las características de las métricas habituales para valorar a los regresores con los que trabajaremos, dado que en estos casos nos centraremos en el error existente respecto a nuestra salida "y" y la predicción hecha del modelo "y'". Este error dado por instancia ocasionalmente subestimará en este caso el 'price' de la propiedad y en otras instancias lo sobreestimará. De modo que consecuentemente tendremos valores positivos y negativos a ese fin. Simplemente sumarlos nos haría caer en el error de 'compensar' valores positivos con negativos y podria darnos una *falsa* idea de que nuestros modelos están bien ajustados. 
##### Una alternativa a eso sería realizar una suma de los valores absolutos (convertir en 'positivos' a nuestros negativos y evitar compensaciones) para luego obtener su media. A esto lo llamamos MAE (error absoluto promedio). Contamos asimismo con la MSE (o error cuadrado promedio) que consiste en potenciar los valores al cuadrado para que a la postre sean positivos, sumarlos y obtener una media de los mismos. Su desventaja reside en la diferencia de magnitudes a causa de la potencia respecto de los *verdaderos* valores de 'y'. Asi entonces tambien aparece en escena RMSE (o raiz del error cuadrado medio), que consiste en aplicarle al MSE una raíz cuadrada.
##### Por lo que comentábamos respecto a la desventaja de las magnitudes de los errores expresados por MSE nos concentraremos ahora en elegir una métrica adecuada entre MAE y RMSE, teniendo siempre presente que nuestra intención en la obtención de rangos que corren desde 0 a ∞ es que podamos dar con resultados lo más cercanos a 0 (cero) que nos sea posible, esto dado que nuestra medición justiprecia errores y a menos errores mejor tendería a funcionar el modelo regresor. Entiendo asi que elevar los errores al cuadrado antes de promediarlos se sintetiza en que RMSE otorgará un *peso relativo* más alto a los errores grandes o extremos de nuestras mediciones respecto a MAE, lo que se traduce en que RMSE es más útil cuando estos errores extremos o grandes son indeseables, demostrando una gran sensibilidad a la varianza de nuestros errores. Desde este punto de vista entiendo que es conveniente usar RMSE en la medida que contamos en nuestro dataset con valores poco uniformes (es decir, con una gran varianza).

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)

#linear_model será nuestro 'benchmark'
linear_model = LinearRegression()
tree = DecisionTreeRegressor(max_depth=12)
knn = KNeighborsRegressor(n_neighbors=5)
```

```python
linear_model.fit(X_train, y_train)
tree.fit(X_train, y_train)
knn.fit(X_train, y_train)
```

##### A continuación predeciremos los resultados de entrenamiento y testeo, para poder mediante un calculo iterado de RMSE (Distancia media cuadrática mínima) sobre train y test evaluar el error de nuestros modelos, con el propósito último de determinar cuál de estos de adapta mejor al objetivo propuesto al comienzo de este trabajo. Graficaremos con un histograma estos errores, esperando ver una forma de campana 'gaussiana' cercana a 0 (cero).

```python
modelos = ['Regresión lineal', 'Árbol de Decisión', 'Vecinos más cercanos']

for i, model in enumerate([linear_model, tree, knn]):
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    print(f'Modelo: {modelos[i]}')
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    print(f'Raíz del error cuadrático medio en Train: {rmse_train}')
    print(f'Raíz del error cuadrático medio en Test: {rmse_test}')
    
    plt.figure(figsize = (20,10))

    plt.subplot(1,2,1)
    sns.distplot(y_train - y_train_pred, bins = 50, label = 'train')
    sns.distplot(y_test - y_test_pred, bins = 50, label = 'test')
    plt.xlabel('errores')
    plt.legend()

    ax = plt.subplot(1,2,2)
    ax.scatter(y_test,y_test_pred, s =2)    
    lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes]
    ]
    
    ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
    plt.xlabel('y (test)')
    plt.ylabel('y_pred (test)')
    
    plt.tight_layout()
    plt.show()
```
